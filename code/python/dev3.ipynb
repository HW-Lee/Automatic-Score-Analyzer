```python
from IPython.display import Image
import matplotlib.pyplot as plot
from scipy import misc
import numpy as np

path = '../../data/score001.jpg'
Image(filename=path)
```

```python
image = misc.imread(name=path, flatten=True)

col_id = round( len(image[0])/2 )
black_idx = list(zip( image[:, col_id], np.r_[0:len(image[:, col_id])] ))
black_idx = list(filter( lambda x: x[0] < 20, black_idx ))
black_idx = list(map( lambda x: x[1], black_idx ))
black_idx
```
