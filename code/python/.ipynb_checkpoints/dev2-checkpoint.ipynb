```python
import json
import numpy as np
import ScoreAnalyzer as sa

data = sa.load("../../data/score011.jpg", thresh=.2)
info = json.load(open("../../data/centers011.json"))
centers = info["centers"]
```

```python
plt.gcf().clear()
for i, center in enumerate(centers):
    plt.plot([i]*len(center), center, 'r.')
    
plt.imshow(data, cmap=plt.cm.binary)

plt.gcf().set_size_inches(plt.gcf().get_size_inches()*6)
plt.gcf().set_dpi(300)
plt.show()
```

```python
cnt = 0
for i, center in enumerate(centers):
    cnt += len(center)

cnt
```

```python
staves = []
height = 5*info["width"] + 4*info["space"]
height
```

```python
staves = sa.find_lines_RANSAC(centers, img_width=data.shape[1], staffline_height=height, max_iter=500)
```

```python
staves
```

```python
pts = sa._get_pts_set(centers)

xx = np.linspace(0, data.shape[1])
for a, b in staves:
    plt.plot(xx, xx*a + b, 'r')
    
plt.plot(pts[:, 0], pts[:, 1], 'b.')
plt.gcf().set_size_inches(12, 9)
plt.show()
plt.gcf().clear()
```

```python
xx = np.linspace(0, data.shape[1])
for a, b in staves:
    plt.plot(xx, xx*a + b, 'r')
    
plt.imshow(1-data, cmap=plt.cm.binary)
plt.gcf().set_size_inches(plt.gcf().get_size_inches()*6)
plt.gcf().set_dpi(300)
plt.show()
plt.gcf().clear()
```

```python
a, b = staves[0]
c = np.floor(data.shape[1] / 2.)
y = int(round(a*c + b))

theta = np.arctan(a)
staff_seg = misc.imrotate(data[y-2*height:y+2*height, :], theta/np.pi * 180., interp="bilinear")
staff_seg = sa._binarize_image(staff_seg, thresh=.2)
```

```python
spacing = info["width"]/2. + info["space"]
for x in range(5):
    plt.plot([0, staff_seg.shape[1]], [staff_seg.shape[0]/2.+(x-2)*spacing] * 2, "r-")

plt.imshow(1-staff_seg, cmap=plt.cm.binary)
plt.gcf().set_size_inches(plt.gcf().get_size_inches()*6)
plt.gcf().set_dpi(300)
plt.show()
plt.gcf().clear()
```
